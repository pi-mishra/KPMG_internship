```python
# defining funtion to modify dataframe for analysis

def row_to_column(df):
    # Assign the first row as column names
    df.columns = df.iloc[0, :]

    # Drop the first row
    df = df.drop(0).reset_index(drop=True)

    # Remove the column names' index (name=0)
    df.columns.name = None

    return df  # Return the modified DataFrame
```

```python
# modifying the dataframe

Transactions = row_to_column(Transactions)
```

```python
# defining a function to get the datatype, null and duplicates

def dataframe_info(df):
    # Get data types of columns
    print("DATA TYPES")
    print(df.dtypes)
    print("-"*50)

    # Get the count of null values in each column
    print("Null Value Sum")
    print(df.isnull().sum())
    print("-"*50)

    # Get the count of duplicate rows
    print(f"Duplicate Rows = {df.duplicated().sum()}")

    return
```

```python
# Getting the datatype , null and duplicates

dataframe_info(Transactions)
```
