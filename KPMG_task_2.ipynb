```python
# defining funtion to modify dataframe for analysis

def row_to_column(df):
    # Assign the first row as column names
    df.columns = df.iloc[0, :]

    # Drop the first row
    df = df.drop(0).reset_index(drop=True)

    # Remove the column names' index (name=0)
    df.columns.name = None

    return df  # Return the modified DataFrame
```

```python
# modifying the dataframe

Transactions = row_to_column(Transactions)
```

```python
# working on null values

Transactions.isnull().mean()*100
```

```python
# checking the null online_order

Transactions[Transactions['online_order'].isnull()]
```

```python
# Impute "Not Available" to NaN values in the "column_name" column

Transactions['online_order'].fillna("Not Available", inplace=True)
```

```python
# Checking nan values of brand

Transactions[Transactions['brand'].isnull()]
```

```python
# we found that brand nan values have some relationship with product_line, product_class,product_size,standard_cost and product_first_sold_date
# if the brand is null other mentioned columns are also null, it might be due to some error. So we are dropping the null rows

# Drop rows with missing values
Transactions.dropna(inplace=True)
```

```python
# converting to datetime

Transactions['transaction_date'] = pd.to_datetime(Transactions['transaction_date'])
Transactions['product_first_sold_date'] = pd.to_datetime(Transactions['product_first_sold_date'])

# converting to float

Transactions['list_price'] = pd.to_numeric(Transactions['list_price'])
Transactions['standard_cost'] = pd.to_numeric(Transactions['standard_cost'])
```
